```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://www.litres.ru/popular/'
website_url = requests.get(link)
```

```python
website_url.status_code
```

```python
soup = BeautifulSoup(website_url.text, 'lxml') #–æ—Ç–¥–µ–ª–µ–Ω–∏–µ —Å–ª—É–∂–µ–±–Ω–æ–π –æ—Ç –Ω–µ—Å–ª—É–∂–µ–±–Ω–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏
soup
```

```python
soup.prettify()
```

```python
dict_df = {
    'title': [],
    'author': [],
    'rating': []
         }
```

```python
list_books = soup.find_all('div', attrs= {'class': \
                                          ['ArtsGrid-module__artWrapper_1j1xJ', \
                                           'ArtInfo-modules__author_235kc']})
list_books
```

```python
list_books[0]
```

```python
print(list_books[0].find('p', class_='ArtInfo-modules__title_2T1yc'))  #–ø—Ä–æ–≤–µ—Ä—è—é, —á—Ç–æ –≤—ã–≤–æ–¥–∏—Ç
```

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
for book in list_books:
    dict_df['title'].append(book.find('p', class_='ArtInfo-modules__title_2T1yc').get_text(strip=True))
    dict_df['author'].append(book.find('a', class_='ArtInfo-modules__author_235kc').get_text(strip=True))
    dict_df['rating'].append(book.find('div', class_='ArtRating-module__rating_5dOun').get_text(strip=True))
```
