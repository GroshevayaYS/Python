```python
# Посчитаем для каждого пользователя:
# количество сделанных звонков и израсходованных минут разговора по месяцам
```

```python
# выделяем месяц и год из даты
df_calls['year_month'] = df_calls['call_date'].dt.to_period('M')
```

```python
df_calls['duration'] = np.ceil(df_calls['duration'])
```

```python
# Группируемся по столбцам 'user_id', 'year_month' и считаем кол-во звонков по 'id'
# и суммируем по 'duration'
df_user_month = df_calls.groupby(['user_id', 'year_month'])\
                .agg(count_call_user_month = ('id', 'count'),
                  sum_duration_user_month = ('duration', 'sum')).reset_index()
```

```python
#вариант решения с помощью pivot_table
pd.pivot_table(data = df_calls,
               index = 'user_id' ,
               columns = 'year_month',
               values = ['id', 'duration'],
               aggfunc = {'id': 'count', 'duration': 'sum'}).fillna(0)
# в конеце меняем пустые значения на 0
# данные пропуски мы видим в месяцах, предшествующих дате заключения договора
```

```python
# обединение с помощью метода '.merge'
df_user_month.head(2)
```

```python
# добавим в df_users датафрем df_user_month с указанием необходимых признаков
# присвоим полученный результат переменной df
df_calls_month = df_users[['user_id', 'first_name', 'last_name', 'city', 'tariff']]\
             .merge(df_user_month[['user_id','count_call_user_month',\
            'sum_duration_user_month','year_month']],
             on = 'user_id',
             how = 'inner')
```

```python
# Чтобы не потерять клиентов, которые могли не пользоваться
# услугами интернета и отправкой сообщений, будем использовать внутренний джоин
```

```python
df_messages['year_month'] = df_messages['message_date'].dt.to_period('M')  # выделим из даты месяц в формате 'год-месяц'
df_messages
```

```python
# Количество отправленных сообщений по месяцам
df_messages_month = df_messages.groupby(['user_id', 'year_month'])\
.agg(count_messages = ('id', 'count'))\
.reset_index()
df_messages_month
```

```python
df_internet['year_month'] = df_internet['session_date'].dt.to_period('M')
df_internet
```

```python
df_internet_month = df_internet.groupby(['user_id', 'year_month'])\
.agg(volume_traffic = ('mb_used', 'sum')).reset_index()
df_internet_month
```

```python
df_call_messages_internet_agg = df_user_month[['user_id', 'year_month', 'count_call_user_month', 'sum_duration_user_month']]\
                                .merge(df_messages_month[['user_id', 'year_month', 'count_messages']],
                                                     on=('user_id', 'year_month'), how='outer')\
                                .merge(df_internet_month[['user_id', 'year_month', 'volume_traffic']],
                                                     on=('user_id', 'year_month'), how='outer')
df_call_messages_internet_agg
```

```python
def month_calc(name_df):
    calls = name_df['sum_duration_user_month']
    messages = name_df['count_messages']
    internet = name_df['volume_traffic']
    tariff = name_df['tariff']
    # достанем лимиты по тарифам на звонки, смс и интернет
    minutes_lim = df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'minutes_included'].iloc[0]
    messages_lim = df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'messages_included'].iloc[0]
    internet_lim = df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'mb_per_month_included'].iloc[0]
    # переплата по звонкам
    call_overlimit = max(calls - minutes_lim, 0)
    call_overpay = call_overlimit * df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'rub_per_minute'].iloc[0]
    # # переплата по сообщениям
    messages_overlimit = max(messages - messages_lim, 0)
    messages_overpay = messages_overlimit * df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'rub_per_message'].iloc[0]
    # # переплата за трафик интернета
    internet_overlimit = max(internet - internet_lim, 0)
    internet_overpay = np.ceil(internet_overlimit/1024)*df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'rub_per_gb'].iloc[0]
    return (call_overpay + messages_overpay + internet_overpay+\
           df_tariffs.loc[df_tariffs['tariff_name'] == tariff, 'rub_monthly_fee'].iloc[0])
```

```python
df['monthly_profit_each_user'] = df.apply(month_calc, axis=1)
```
